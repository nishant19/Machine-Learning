### Titanic

```python
titanic = pd.read_csv('titanic_train.csv')
titanic.head()
del titanic['PassengerId']
del titanic['Name']
del titanic['Ticket']
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
del titanic['Cabin']
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.Age.fillna(titanic.Age.mean(), inplace = True)
titanic.describe()
```

```python
survived = titanic[titanic.Survived == 1]
survived.describe()
didntsurvive = titanic[titanic.Survived == 0]
didntsurvive.describe()
```

```python
survivedS = titanic[titanic.Embarked == 'Q'][titanic.Survived == 1].shape[0]
survivedS
```

```python
titanic.dropna(inplace = True)
titanic.shape
```

```python
def getNumber(str):
    if (str == 'male'):
        return 1
    else :
        return 2

titanic['gender'] = titanic['Sex'].apply(getNumber)
titanic.head()
del titanic['Sex']
```

```python
titanic.head()
```

```python
males = (titanic['gender'] == 1).sum()
females = (titanic['gender'] == 2).sum()

p = [males, females]
plt.pie(p,
       labels = ['Male', 'Female'],
       colors = ['blue', 'red'],
       explode = (0.15, 0),
       startangle = 0)
plt.axis('equal')
plt.show()
```
